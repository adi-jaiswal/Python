## 1. Create a function in python to read the text file and replace specific content of the file.


```python
def modify_file(path, replace_content, replace_with):
    with open(path, "r") as f:
            for i in f:
                l = str(i).split(' ')
            j = l.index(replace_content)
            l[j] = replace_with
            newData = ' '.join(l)
            
    with open(path, "w") as f:
        f.write(newData)
    
    with open(path, "r") as f:
        print(f.read())
```

```python
modify_file('example.txt', replace_content = 'placement', replace_with = 'screening')
```

### Abstract Classes

An abstract class is a blueprint for other classes. It allows us to create a set of methods that must be created within any child classes built from the abstract class. A class which contains one or more abstract methods is called an abstract class.

```python
from abc import ABC, abstractmethod

class laptop(ABC):
    @abstractmethod
    def processor(self):
        pass
    
class dell(laptop):
    def processor(self):
        print("The processor is i7")
        
class acer(laptop):
    def processor(self):
        print("The processor is i5")
    
class hp(laptop):
    def processor(self):
        print("The processor is i3")

a = dell()
a.processor()

b = acer()
b.processor()

c = hp()
c.processor()
```

```python
from abc import ABC, abstractmethod

class smartphone(ABC):
    @abstractmethod
    def camera(self):
        pass
    
class iphone13(smartphone):
    def camera(self):
        print("The camera of iPhone 13 is 12 megapixels")
        
class samsungs22(smartphone):
    def camera(self):
        print("The camera of Samsung S22 is 108 megapixels")
    
class oneplus10(smartphone):
    def camera(self):
        print("The camera of OnePlus 10 is 48 megapixels")

a = iphone13()
a.camera()

b = samsungs22()
b.camera()

c = oneplus10()
c.camera()
```

### Mutiple Inheritence

When a class is derived from more than one base class it is called multiple Inheritance. The derived class inherits all the features of the base class.

```python
#Multiclass Inheritence

class xyz:
    
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
    
    def test(self):
        print("This is a test method of class xyz")
        
class xyz1:
    
    def __init__(self,p,q,r):
        self.p = p
        self.q = q
        self.r = r
        
    def test1(self):
        print("This is a test1 method of class xyz1")

class child(xyz1,xyz):
    
    def __init__(self,*args,**kwargs):
        xyz.__init__(self,*args)
        xyz1.__init__(self,**kwargs)
```

```python
n = child(4,5,6,p=1,q=2,r=3)
```

```python
n.a
```

```python
n.p
```

```python
#Multilevel Inheritence

class xyz:
    
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
    
    def test(self):
        print("This is a test method of class xyz")
        
class xyz1(xyz):
        
    def test1(self):
        print("This is a test1 method of class xyz1")
        
class xyz2(xyz1):
    
    def test2(self):
        print("This is test2 method of class xyz2")

class xyz3(xyz2):
    
    def test3(self):
        print("This is test3 method of class xyz3") 
```

```python
v = xyz3(45,6,67)
```

```python
v.test3()
```

### Decorator

Decorators allow us to wrap another function in order to extend the behaviour of the wrapped function, without permanently modifying it. 

```python
def test(a):
    return a.upper()
 
print(test('Hi there!'))
 
test1 = test
 
print(test1('HI there!'))
```
